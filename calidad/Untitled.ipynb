```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
my_list = [2,4,6,8,10]
my_array = np.array(my_list)

h = []
with open('QMYC3direct1Mock.txt', 'r') as toma:
    for line in toma:
        h.append(int(line))
# my_array = np.array(lista)
# my_array = np.sort(my_array)
h.sort()
hmean = np.mean(h)
hstd = np.std(h)
pdf = stats.norm.pdf(h, hmean, hstd)
plt.plot(h, pdf)
```
